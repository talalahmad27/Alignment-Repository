```python
import openai
import os
import random
import pandas as pd
from openai import OpenAI
client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key="",
)

data_frame = pd.read_excel('Dataset_Misinformation.xlsx')

final = {
    'Question': [],  # Placeholder for the first column
    'Rejected Response': []   # Placeholder for the second column
}

# Create the DataFrame

for i in range(0,len(data_frame)):
    text =  data_frame['Misinformation'].iloc[i]
    prompt = (
            f'The following text contains some misinformation regarding the vaccine. Write me a probable question that is relevant to this'
            f'below pasted misinformation:'
            f'{text}'
        )
    response = client.chat.completions.create(
            model="gpt-4o-mini",
            temperature=0.8,
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=200
        )
    final['Question'].append(response.choices[0].message.content.strip())
    final['Rejected Response'].append(text)

df_final = pd.DataFrame(final)
df_final.head(10)

```

```python
df_final.to_csv('./output_misinformation_questions.csv', index=False)
from google.colab import files
files.download('./output_misinformation_questions.csv')
```
